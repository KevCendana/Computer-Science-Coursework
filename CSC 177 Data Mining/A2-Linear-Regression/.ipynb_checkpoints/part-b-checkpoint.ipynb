```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import re
data = pd.read_csv("Admission_Predict_Ver1.1_small_data_set_for_Linear_Regression.csv", encoding="latin")
data.columns # identify columns to use
```

### Train-Test Split

```python
# import function to split data into training and testing sets
from sklearn.model_selection import train_test_split 

# identify columns to use
features = data[['GRE Score', 'CGPA']] 
target = data['University Rating']

# split data into training and testing sets (~ %80 / %20)
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

### Standardizing Data

```python
from sklearn.preprocessing import StandardScaler # import function to scale data

# use the same scaler for all test data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Simple Linear Regression using 2 variables: GRE Score + University Rating

```python
# import linear regression and plotting library
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# label testing and training data for simple regression with 'GRE Score' 
X_simple_train = X_train_scaled[:, 0].reshape(-1, 1)  # GRE Score
X_simple_test = X_test_scaled[:, 0].reshape(-1, 1)

# train simple linear regression model, then predict the values
simple_reg = LinearRegression().fit(X_simple_train, y_train)
y_pred_simple = simple_reg.predict(X_simple_test)

# plot actual vs predicted values 
plt.scatter(X_simple_test, y_test, color='blue')
plt.plot(X_simple_test, y_pred_simple, color='red')
plt.title('GRE Score vs. University Rating')
plt.xlabel('Standardized GRE Score')
plt.ylabel('University Rating')
plt.show()
```
